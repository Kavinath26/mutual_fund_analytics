```python
# %% [cell 1] Load and Explore Fund Master (Task 6)
import pandas as pd
import os

# Define paths
raw_dir = "../data/raw"
master_path = os.path.join(raw_dir, "01_fund_master.csv")
nav_history_path = os.path.join(raw_dir, "02_nav_history.csv")

# Read datasets
fund_master = pd.read_csv(master_path)
nav_history = pd.read_csv(nav_history_path)

print("--- UNIQUE FUND HOUSES ---")
print(fund_master['fund_house'].unique() if 'fund_house' in fund_master.columns else "Column name differs")

print("\n--- UNIQUE CATEGORIES ---")
print(fund_master['category'].unique() if 'category' in fund_master.columns else "Column name differs")

print("\n--- UNIQUE SUB-CATEGORIES ---")
print(fund_master['sub_category'].unique() if 'sub_category' in fund_master.columns else "Column name differs")

print("\n--- RISK GRADES DISTRIBUTION ---")
if 'risk_grade' in fund_master.columns:
    print(fund_master['risk_grade'].value_counts())
elif 'risk_level' in fund_master.columns:
    print(fund_master['risk_level'].value_counts())
```

```python
# %% [cell 2] AMFI Code Cross-Validation (Task 7)
# Dynamically find the primary key column for scheme codes
master_code_col = [col for col in fund_master.columns if 'code' in col or 'scheme' in col][0]
nav_code_col = [col for col in nav_history.columns if 'code' in col or 'scheme' in col][0]

master_codes = set(fund_master[master_code_col].dropna().unique())
nav_codes = set(nav_history[nav_code_col].dropna().unique())

# Identify codes present in master but missing in history
missing_codes = master_codes - nav_codes

print("=== DATA QUALITY REPORT ===")
print(f"Total Unique Schemes in Fund Master: {len(master_codes)}")
print(f"Total Unique Schemes in NAV History: {len(nav_codes)}")
print(f"Schemes in Master missing from historical logs: {len(missing_codes)}")

if len(missing_codes) == 0:
    print("\nSUCCESS: Data validation passed. Integrity holds perfectly.")
else:
    print(f"\nWARNING: Integrity gap identified. {len(missing_codes)} schemes lack historical NAV data.")
    print(f"Sample mismatched AMFI codes: {list(missing_codes)[:5]}")
```
